This notebook generates a list of my GitHub repos in MediaWiki format.
Presently, the list is written to a text file, **repo-list.txt**.
Contents of this file can be copy and pasted over the contents of https://www.guaminsects.net/aubwiki2020/index.php?title=GitHub_Repositories. AUTOMATION NEEDED HERE.

Please remember to update https://github.com/aubreymoore/miscellaneous-jupyter-notebooks/blob/master/repo-list.ipynb whenever this notebook is updated. AUTOMATION NEEDED HERE.

```python
from github import Github
import getpass
#from mwclient import Site
```

```python
github_repo_url = 'https://github.com/aubreymoore/miscellaneous-jupyter-notebooks'
notebook_filename = 'repo-list.ipynb'
github_url = '{}/blob/master/{}'.format(github_repo_url, notebook_filename)
```

```python
password = getpass.getpass()
g = Github("aubreymoore", password)
```

```python
s = 'This list was generated by a [{}, Jupyter Notebook].\n'.format(github_url)
s += '= List of My GitHub Repositories =\n'
for repo in g.get_user().get_repos():
    s += '#[{} {}]\n'.format(repo.html_url, repo.name)
with open('repo-list.txt', 'w') as f:
    f.write(s)
```
